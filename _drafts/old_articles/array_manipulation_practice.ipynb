```python
# right rotation of array
def solution(A, K):
    
    n = len(A)
    if n == 0:
        return A
    
    A_ = list(A)
    K_ = K % n
    
    left_array = A[-K_:] if K_ > 0 else []
    right_array = A[:n - K_] if K_ > 0 else A_
    
    A_ = left_array + right_array
    
    return A_
    
```

```python
# seen elements
def solution(X, A):
    
    leaves_pos = dict.fromkeys(range(1, X+1), 0)
    
    for (sec, leave_pos) in enumerate(A):
        
        # if first leaf in this position, we remove it from dict
        if leave_pos in leaves_pos:
            del leaves_pos[leave_pos]
            
        # if all positions have been filled, our dict is empty
        if len(leaves_pos) == 0:
            return sec
            
    return -1
```

```python
# permutation array (solution 1)
def solution(A):
    
    # array of all possible integers
    N = len(A)
    values_count = [0] * (N+1)
    
    # count how many times we see a value
    for k in range(N):
        # if the integer is bigger than the array length, it cannot be a permutation
        if A[k] > N:
            return 0
        
        # otherwise, we count the new occurence
        values_count[A[k]] += 1
        
    
    # we want only one occurrence per value
    for k in range(1, N+1):
        if values_count[k] != 1:
            return 0
            
    return 1
    
```

```python
# permutation array (solution 2)
def solution(A):
    
    N = len(A)
    int_max = max(A)
    
    # the max value must be equal to N
    if int_max != N:
        return 0
        
    # list of expected values
    values_count = dict.fromkeys(range(1, N+1), 0)
    
    for num in A:
        # first time we see a value: we remove it from the dict
        if num in values_count:
            del values_count[num]
    
        # second time we see a value: not a permutation
        else:
            return 0

    # we have only seen each value once and int_max == N, it's a permutation
    return 1
```

```python
# smaller unseen positive integer
def solution(A):

    # max integer
    max_int = max(A)
    
    # only negative integers, we return 1
    if max_int <= 0:
        return 1
        
    # array of all possible integers
    values_count = [0] * (max_int+1)
    
    # count how many times we see a positive value
    N = len(A)
    for k in range(N):
        if A[k] > 0:
            values_count[A[k]] += 1 # we only count positive integers. negative ones would count from the end
        
    # we want the first value with zero occurrences
    for k in range(1, max_int+1):
        if values_count[k] == 0:
            return k
            
    # we have seen all numbers up until max_int, so we return max_int + 1
    return max_int + 1
```

```python
# counters increment
def solution(N, A):
    
    counters = [0] * N
    
    # max counter value
    max_counter = 0
    
    # we check if the previous operation was a max counter (if the new one is also a max, then nothing changes)
    is_prev_max = 0
    
    for num in A:
        
        if num == N+1:
            # we do something only if the previous operation was not max_counter
            if is_prev_max == 0:
                # we update the is_prev_max flag 
                is_prev_max = 1
                
                # we increase all our couter values
                counters = [max_counter] * N
                
        else:
            # we update the is_prev_max flag 
            is_prev_max = 0
            
            # we increase our counter 1 the max value if needed
            counters[num-1] += 1
            if counters[num-1] > max_counter:
                max_counter = counters[num-1]
                
    return counters
```

```python
#passing cars
def solution(A):
    
    # we sum the number of cars going east
    n = len(A)
    P = [0] * (n + 1)
    
    passing_cars = 0
    
    for k in range(1, n + 1):
        # car is going east
        if A[k - 1] == 0:
            P[k] = P[k - 1] + 1
            
        # car is going west and will pass all the previous cars going east
        else:
            P[k] = P[k - 1]
            passing_cars += P[k]
            
            # special case: too many passing cars
            if passing_cars > 1e9:
                return -1
    
    # print(P)
    return passing_cars
```

```python
# min nucleotide

def count_letters(S, letter):
    # count the number of times we have seen a nucleotide
    
    N = len(S)
    
    # init array
    letter_counts = [0] * (N + 1)
    
    for k in range(1, N+1):
        if S[k-1] == letter:
            letter_counts[k] = letter_counts[k-1] + 1
        else:
            letter_counts[k] = letter_counts[k-1]
        
    return letter_counts
    
    
def is_letter_in_slice(letter_counts, P, Q):
    # count the occurrences of a given letter between P and Q
    slice_occurrences = letter_counts[Q+1] - letter_counts[P]
    return slice_occurrences > 0
    
    
def solution(S, P, Q):
    
    # count the number of times we have seen a nucleotide
    A_counts = count_letters(S, 'A')
    C_counts = count_letters(S, 'C')
    G_counts = count_letters(S, 'G')

    # init results
    M = len(P)
    results = [0] * M
    
    for k in range(M):
        P_k = P[k]
        Q_k = Q[k]

        # we check the occurrences of each letter
        if is_letter_in_slice(A_counts, P_k, Q_k):
            results[k] = 1
        elif is_letter_in_slice(C_counts, P_k, Q_k):
            results[k] = 2
        elif is_letter_in_slice(G_counts, P_k, Q_k):
            results[k] = 3
        else:
            results[k] = 4
        
    return results
        
```

```python
# triangular triplets
def solution(A):
    
    # let's assume C > B > A
    # A + B > C   # to check
    # B + C > A   # always true
    # C + A > B   # always true

    # note: all numbers must be > 0
    
    A_ = sorted(A)
    
    N = len(A_)
    for k in range(N - 2):
        if (A_[k] > 0 and A_[k] + A_[k+1] > A_[k+2]):
            return 1
            
    return 0
```

```python
# biggest triplet product
def solution(A):
   
    A_ = sorted(A)
    N = len(A_)     # N >=3
    
    # same sign for all elements: we multiply the biggest values
    if (A_[0] >= 0) or (A_[-1] <= 0):
        return A_[-3] * A_[-2] * A_[-1]
    
    # otherwise, at least one positive integer.
    # we may have two large negative numbers to take into account
    else:
        return A_[-1] * max(A_[-3] * A_[-2], A_[0] * A_[1])
        
```

```python
# stack
stack = [0] * N
size = 0

def push(x):
    global size
    stack[size] = x
    size += 1

def pop():
    global size
    size -= 1
    return stack[size]
```

```python
# queue
queue = [0] * N
head, tail = 0, 0

def push(x):
    global tail
    tail = (tail + 1) % N
    queue[tail] = x

def pop():
    global head
    head = (head + 1) % N
    return queue[head]

def size():
    return (tail - head + N) % N

def empty():
    return head == tail
```

```python
# fishes
def solution(A, B):
    # we save the downstream fishes as we see them
    # if upstream fish, we check which downstream fishes survive it
    # LIFO stack

    N = len(A)
    
    surviving = 0
    downstream_fish = []

    for k, fish_size, fish_dir in zip(range(N), A, B):
        
        # downstream fishes
        if fish_dir == 1:
            downstream_fish.append(fish_size)
            
        # upstream fishes
        else:
            while downstream_fish and fish_size > downstream_fish[-1]:
                downstream_fish.pop()
        
            # check if while loop stopped because no more fish or bigger
            if downstream_fish and downstream_fish[-1] > fish_size:
                continue
            
            else:
                surviving += 1
                
    
    surviving += len(downstream_fish)
    return surviving
    
```

```python
# walls
def solution(H):
    # we save the height of the current walls 
    # if current_wall higher than next height, we need a new wall
    # LIFO stack

    N = len(H)
    
    total_walls = 0
    current_heights = []
    max_current_height = 0
    
    for height in H:
        
        # close existing walls that are too big for the new needs
        while current_heights and height < current_heights[-1]:
            current_heights.pop()
            total_walls += 1
                
        # existing wall covers needs
        if current_heights and height == current_heights[-1]:
            continue
        
        # existing walls too small or no existing wall
        else:
            current_heights.append(height)

    total_walls += len(current_heights)
    
    return total_walls
```

## Sieve of Erathosthene

```python
def sieve(n):

    # init seave (0 .. N)
    sieve = [True] * (n + 1)

    # 0 and 1 are not prime
    sieve[0] = sieve[1] = False
    
    # loop until sqrt(n)
    i=2
    while (i * i <= n):
        
        # if number is prime, we remove all its multiple
        if (sieve[i]):

            # we start at i² because if k < i² then k * i is also a multiple of k than were already removed
            k = i * i
            while (k <= n):
                sieve[k] = False
                k += i

        i += 1

    return sieve
```

```python
# smallest prime divisor of numbers in range 0..N
def arrayF(n):

    F = [0] * (n + 1)

    i=2
    while (i * i <= n):

        # if number is prime
        if (F[i] == 0):

            # loop from i²
            k = i * i
            while (k <= n):
                
                # F[k] has no smaller divisor, we add our prime number
                if (F[k] == 0):
                    F[k] = i
                
                # next multiple of i
                k += i
        
        # next number
        i += 1
    
    return F

# factorization of x
def factorization(x, F):

    primeFactors = []

    # we loop over all prime factors
    while (F[x] > 0):
        
        # we add the prime factor to the array
        primeFactors.append(F[x])

        # then check if x / F[x] has a prime factor
        x = x // F[x]
    
    # no more prime factors: the remaining x is prime
    primeFactors.append(x)
    
    return primeFactors
```

## Fibonacci


```python
def fibonacciDynamic(n):
    fib = [0] * (n + 2)
    fib[1] = 1
    for i in range(2, n + 1):
        fib[i] = fib[i - 1] + fib[i - 2]
    
    return fib[n]
```

```python
# frog
def fibonacci(n):
    fib = [0] * (n + 2)
    fib[1] = 1
    for i in range(2, n + 1):
        fib[i] = fib[i - 1] + fib[i - 2]
    
    return fib[n]
    
    
def solution(A):

    N = len(A)
    
    # possible moves
    fib = []
    fib_k = 0
    while fib_k < max(N, 3):
        fib_k = fibonacci(len(fib))
        fib.append(fib_k)

    # greedy moves
    i = -1
    jump_count = 0
    
    while i < N:
        
        i_tmp = i
        for k in range(len(fib) - 1, 0, -1):
            
            # too big
            if i + fib[k] > N:
                continue
            
            # in there
            elif i + fib[k] == N:
                jump_count += 1
                i+= fib[k]
                break
            
            elif A[i + fib[k]] == 1:
                jump_count += 1
                i+= fib[k]
                break
            
        # no possible jump
        if i == i_tmp:
            return -1
        
    return jump_count
```

## Binary Search


```python
def binarySearch(A, x):
    '''
    returns index of largest element of a sorted array A that is <= x
    '''

    # two extremities of the array
    n = len(A)
    beg = 0
    end = n - 1
    
    # largest index
    result_index = -1

    while (beg <= end):

        # split array in two
        mid = (beg + end) // 2

        # if mid matches the condition: 
        # update result and keep looking in right half
        if (A[mid] <= x):
            result_index = mid
            beg = mid + 1
            
        # keep left half otherwise
        else:
            end = mid - 1

    return result_index

binarySearch([1, 2, 3], 2)
```

```python
def boards(A, k):

    n = len(A)
    beg = 1
    end = n
    result = -1

    while (beg <= end):
        mid = (beg + end) / 2
        if (check(A, mid) <= k):
            end = mid - 1
            result = mid
        else:
            beg = mid + 1

    return result


def check(A, k):
    
    n = len(A)
    boards = 0
    last = -1

    for i in range(n):
        if A[i] == 1 and last < i:
            boards += 1
            last = i + k - 1

    return boards
```

```python
#min max
def blockSizeIsValid(A, K, max_sum_candidate):
    '''
    check if we can find a block 
    '''
    
    block_sum = 0
    block_cnt = 0
 
    for element in A:
        
        # adding a new element would be above max_sum_candidate, we create a new one
        if block_sum + element > max_sum_candidate:
            
            block_cnt += 1       # closing existing block
            block_sum = element  # initialize the new block
        
        # otherwise, we continue with the current block   
        else:
            block_sum += element
        
        # we need too many blocks: max_sum_candidate is not feasible
        if block_cnt >= K:
            return False

    return True
 
 
def binarySearch(A, K, M):
    '''
    Binary search for the minimal size of a block
    '''
    
    # min and max values of the sum
    min_sum = M
    max_sum = sum(A)
 
    if K == 1:      return max_sum  # we can only consider the entire array
    if K >= len(A): return min_sum  # we can return each element individually
    
    # binary search on smallest valid sum 
    while min_sum <= max_sum:
        
        mid_sum = (min_sum + max_sum) // 2
        
        # condition
        if blockSizeIsValid(A, K, mid_sum):
            max_sum = mid_sum - 1
            
        # mid_sum is too low: we would require too many blocks to make it true
        else:
            min_sum = mid_sum + 1
 
    return min_sum
 


def solution(K, M, A):
  
    # array of N integers 0 .. N-1 between 0 and M
    # K blocks of different sizes (potentially empty blocks)
    # the large sum is the max sum
    
    # we can have between 1 and K non-empty blocks 
    # each element is positive, so the sum of the entire array is always the max sum
    # it means that the min large sum is for K non-empty blocks.
    
    M = max(A)
    return binarySearch(A, K, M)
    
    
    
```

## Caterpillar

```python
# caterpillar method
def caterpillarMethod(A, s):
    '''
    check if A contains a subsequence of sum s
    '''

    n = len(A)
    front = 0   # first position in subarray
    total = 0   # total of subarray

    for back in range(n):
        
        # increase subarray while not at the end and resulting total <= s
        while (front < n and total + A[front] <= s):
            total += A[front]
            front += 1

        # we found s
        if total == s:
            return True

        # otherwise, our total is now too big: we remove the first number
        total -= A[back]

    return False
```

```python
# count of triangles
def triangles(A):
    '''
    count the number of triplets in a sorted
    array A where (a + b > c)
    '''

    n = len(A)
    result = 0
    
    for x in range(n):

        # y > x
        for y in range(x + 1, n):
            
            z = y + 1

            # continue to check z until the condition becomes untrue (largest stick where x + y > z)
            while (z < n and A[x] + A[y] > A[z]):
                z += 1

            # we found all the z values for (x, y)
            result += z - y - 1
    
    return result
```

```python
# distinct slices
def solution(M, A):
    
    # array A of 0 .. N-1 integers between 0 and M
    # count the number of distinct slices
    
 
    n = len(A)
    front = 0   # first position in subarray
    values_in_slice = {}
    slices = 0
    
    for back in range(n):
        
        # increase subarray while no dupes
        while (front < n and A[front] not in values_in_slice):
            values_in_slice[A[front]] = 0   # add value to dictionary
            slices += 1                     # the current slice is distinct
            front += 1                      # we extend the slice to a new element to check it

        # if dupe: new slice
        front = back + 1
        values_in_slice = {}

    return slices   
    
```

## Greedy Algorithms


```python
def greedyCoinChanging(M, k):
    '''
    Returns the minimum amount of coins to pay the amount k.
    M list all the possible coin values.
    '''

    n = len(M)
    result = []

    # we use the largest denominations first
    for i in range(n - 1, -1, -1):
        result.append([(M[i], k // M[i])]) # value and number of coins
        k %= M[i]

    return result
```

```python
def greedyCanoeistA(W, k):
    '''
    List of canoeists weights W (ordered).
    Maximum load: k.
    Returns the smallest number of canoes.
    '''

    N = len(W)

    # split the canoeists in two groups
    lights = deque()
    heavys = deque()

    for i in range(N - 1):
        # list of canoeists who can sit with the heaviest person
        if W[i] + W[-1] <= k:
            lights.append(W[i])

        # list of persons who cannot
        else:
            heavys.append(W[i])
    
    heavys.append(W[-1])

    # iterate on both lists
    canoes = 0
    while (lights or heavys):

        # new canoe
        canoes += 1

        # heaviest person goes into the canoe
        heavys.pop()
        
        # we can add a light person if they exist (they can sit with the heaviest person)
        if len(lights) > 0:
            lights.pop()

        # if no more heavys, we move the heaviest light 
        if (lights and not heavys):
            heavys.append(lights.pop())

        # we move all the lightest heavys to the lights bucket
        while (len(heavys) > 1 and heavys[0] + heavys[-1] <= k):
            lights.append(heavys.popleft())

    return canoes
```
